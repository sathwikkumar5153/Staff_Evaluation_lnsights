```python
#import required librarires
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#check the employee range based on target column that left
sns.countplot(x='left',data=df)
plt.show()
#in training set less people left , data is imbalance
```

```python
#for salary
sns.countplot(x='salary',data=df , palette=['#FF5733', '#33FF57', '#3357FF'])
plt.show()
#low salary employees are more compared to medium and high
```

```python
#split over salary and left
sns.catplot( x='left' , col='salary', kind='count' , data=df )
#low salaried people are leaving more
```

```python
#for promotion
sns.countplot(x='promotion_last_5years' , data = df)
#promotion range is low
```

```python
#finding out leaving and promotion
sns.catplot(x = 'left' , col='promotion_last_5years' , kind = 'count' , data = df)
```

```python
#checking work accident
sns.countplot(x='Work_accident',data = df)
#have less people who made work accident
```

```python
#checking feature left and work accident
sns.catplot(x='left' , col = 'Work_accident' , kind='count' , data = df)
#people who left did not make any accidents
```

```python
#satisfaction level plot
sns.displot(df['satisfaction_level'])
#value ranges from 0 to 1 and max at 0.1
```

```python
#to show w.r.t to categories
sns.boxplot(x='left' , y = 'satisfaction_level' , data = df)
#employee who left has less satisfation from 0.2 to 0.5
```

```python
sns.displot(df['number_project'])
```
